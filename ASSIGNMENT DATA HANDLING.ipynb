#Q.18.What is the difference between os.fork() and multiprocessing in Python?
#Ans.18.The primary difference between os.fork() and Python's multiprocessing module lies in their level of abstraction and portability.

#os.fork():
#Low-level System Call:
#os.fork() is a direct wrapper around the Unix/Linux fork() system call. It creates a new process (child process) that is an exact copy of the calling process (parent process) at the time of the call.

#Resource Inheritance:
#The child process inherits all resources from the parent, including memory space, open file descriptors, and environment variables.

#Platform Specific:
#os.fork() is only available on POSIX-compliant systems (Unix, Linux, macOS) and is not supported on Windows.

#Manual Management:
#Managing inter-process communication (IPC) and synchronization between parent and child processes requires manual implementation using low-level mechanisms like pipes or shared memory.

#Potential Issues with Multithreading:
#Forking a multithreaded process can lead to issues and is generally discouraged due to potential deadlocks or inconsistent states.

#multiprocessing Module:
#Higher-Level Abstraction:
#The multiprocessing module provides a higher-level, more user-friendly API for creating and managing processes in Python. It abstracts away many of the complexities of low-level process management.

#Portability:
#It offers cross-platform compatibility, working on both POSIX systems and Windows. On Windows, it typically uses a "spawn" method to create new processes, which involves starting a fresh Python interpreter.

#Built-in IPC Mechanisms:
#The module includes convenient tools for inter-process communication and synchronization, such as Queue, Pipe, Lock, and Event, simplifying data sharing and coordination.

#Process Isolation:
#Each process created by multiprocessing has its own independent memory space, ensuring isolation and preventing unintended side effects between processes.

#Recommended for General Use:
#For most parallel programming tasks in Python, the multiprocessing module is the recommended approach due to its ease of use, portability, and robust features for managing processes and communication.


#Q.19. What is the importance of closing a file in Python?

```python
'''#Q.4.Write a Python script that reads from one file and writes its content to another file.'''
#Ans.4.
def copy_file_content(source_file_path, destination_file_path):
    """
    Reads the content from a source file and writes it to a destination file.

    Args:
        source_file_path (str): The path to the file to read from.
        destination_file_path (str): The path to the file to write to.
    """
    try:
        with open(source_file_path, 'r') as source_file:
            content = source_file.read()

        with open(destination_file_path, 'w') as destination_file:
            destination_file.write(content)

        print(f"Content successfully copied from '{source_file_path}' to '{destination_file_path}'.")

    except FileNotFoundError:
        print(f"Error: One of the files was not found. Please check the paths.")
    except IOError as e:
        print(f"Error during file operation: {e}")

# Example usage:
if __name__ == "__main__":
    # Create a dummy source file for demonstration
    with open("source.txt", "w") as f:
        f.write("This is some sample text.\n")
        f.write("This will be copied to another file.")

    source_file = "source.txt"
    destination_file = "destination.txt"
    copy_file_content(source_file, destination_file)
```

```python
'''#Q.6.Write a Python program that logs an error message to a log file when a division by zero exception occurs.'''
#Ans.6.
import logging

# Configure logging
# This sets up the logger to write messages to 'error_log.log'
# The level is set to INFO, meaning INFO, WARNING, ERROR, and CRITICAL messages will be logged.
# The format includes timestamp, log level, and the message.
logging.basicConfig(filename='error_log.log', level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def safe_division(numerator, denominator):
    """
    Performs division and logs an error if ZeroDivisionError occurs.
    """
    try:
        result = numerator / denominator
        logging.info(f"Division successful: {numerator} / {denominator} = {result}")
        return result
    except ZeroDivisionError:
        error_message = f"Attempted division by zero: {numerator} / {denominator}"
        logging.error(error_message) # Log the error message
        print(f"Error: {error_message}. Check 'error_log.log' for details.")
        return None

# Example usage
safe_division(10, 2)
safe_division(5, 0)
safe_division(20, 4)
safe_division(7, 0)
```

```python
'''#Q.8.Write a program to handle a file opening error using exception handling.'''
#Ans.8.
def safe_file_read(filename):
    """
    Attempts to open and read a file, handling potential FileNotFoundError
    and other IOError exceptions.
    """
    try:
        # Attempt to open the file in read mode ('r')
        with open(filename, 'r') as file:
            content = file.read()
            print(f"File '{filename}' content:\n{content}")
    except FileNotFoundError:
        # Handle the specific case where the file does not exist
        print(f"Error: The file '{filename}' was not found.")
    except IOError as e:
        # Handle other input/output errors
        print(f"An I/O error occurred while accessing '{filename}': {e}")
    except Exception as e:
        # Catch any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")
    finally:
        # This block will always execute, regardless of whether an exception occurred
        print("File operation attempt completed.")

# Example usage:
# Case 1: File exists and can be read
print("--- Attempting to read an existing file ---")
with open("my_existing_file.txt", "w") as f:
    f.write("This is a test file.")
safe_file_read("my_existing_file.txt")

print("\n--- Attempting to read a non-existent file ---")
safe_file_read("non_existent_file.txt")

print("\n--- Attempting to read a file with potential permission issues (example) ---")
# On some systems, trying to open a system file for writing without permission
# might raise a PermissionError (a type of IOError).
# This example might not trigger an error on all systems.
safe_file_read("/etc/shadow") # This file typically requires root permissions to read
```

```python
'''#Q.10.How can you append data to an existing file in Python?'''
#Ans.10.
def append_data_to_file(file_path, data_to_append):
    """
    Appends data to an existing file.

    Args:
        file_path (str): The path to the file.
        data_to_append (str): The data to append to the file.
    """
    try:
        with open(file_path, 'a') as file:
            file.write(data_to_append)
        print(f"Data successfully appended to '{file_path}'.")
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError as e:
        print(f"Error appending to file '{file_path}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
file_to_append = "my_append_file.txt"

# Create the file with some initial content
with open(file_to_append, "w") as f:
    f.write("This is the initial content.\n")

# Append new data to the file
append_data_to_file(file_to_append, "This is the appended content.\n")

# Append more data
append_data_to_file(file_to_append, "This is more appended content.")

# Verify the content (optional)
with open(file_to_append, "r") as f:
    print("\nContent after appending:")
    print(f.read())
```

```python
'''#Q.12. Write a program that demonstrates using multiple except blocks to handle different types of exceptions.'''
#Ans.12.
def divide_and_access_list(numerator, denominator, data_list, index):
    """
    Performs division and list access, handling potential exceptions
    with multiple except blocks.
    """
    try:
        result = numerator / denominator
        print(f"Division result: {result}")

        item = data_list[index]
        print(f"Item at index {index}: {item}")

    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
    except IndexError:
        print("Error: Invalid index for the list!")
    except TypeError:
        print("Error: Invalid data types for the operation.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:

# Case 1: No errors
print("--- Case 1: No errors ---")
divide_and_access_list(10, 2, [1, 2, 3], 1)

# Case 2: ZeroDivisionError
print("\n--- Case 2: ZeroDivisionError ---")
divide_and_access_list(10, 0, [1, 2, 3], 1)

# Case 3: IndexError
print("\n--- Case 3: IndexError ---")
divide_and_access_list(10, 2, [1, 2, 3], 5)

# Case 4: TypeError (example with invalid data type for division)
print("\n--- Case 4: TypeError ---")
divide_and_access_list("abc", 2, [1, 2, 3], 1)

# Case 5: Another IndexError
print("\n--- Case 5: Another IndexError ---")
divide_and_access_list(10, 2, [1, 2, 3], -1)
```

```python
'''#Q.15.Write a Python program that prints the content of a file and handles the case when the file is empty.'''
#Ans.15.
import os

def print_file_content(file_path):
    """
    Reads a file and prints its content, handling FileNotFoundError and empty files.
    """
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            if content:
                print(f"Content of '{file_path}':\n{content}")
            else:
                print(f"The file '{file_path}' is empty.")
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError as e:
        print(f"An I/O error occurred while accessing '{file_path}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:

# Case 1: File exists and has content
print("--- Reading a file with content ---")
with open("file_with_content.txt", "w") as f:
    f.write("This file has some text in it.")
print_file_content("file_with_content.txt")

# Case 2: File exists but is empty
print("\n--- Reading an empty file ---")
with open("empty_file.txt", "w") as f:
    pass # Create an empty file
print_file_content("empty_file.txt")

# Case 3: File does not exist
print("\n--- Attempting to read a non-existent file ---")
print_file_content("non_existent_file.txt")

# Clean up the created files (optional)
# os.remove("file_with_content.txt")
# os.remove("empty_file.txt")
```

```python
'''#Q.17.Write a Python program to create and write a list of numbers to a file, one number per line.'''
#Ans.17.
def write_numbers_to_file(file_path, numbers):
    """
    Writes a list of numbers to a file, one number per line.

    Args:
        file_path (str): The path to the file to write to.
        numbers (list): A list of numbers to write.
    """
    try:
        with open(file_path, 'w') as file:
            for number in numbers:
                file.write(str(number) + '\n')
        print(f"Numbers successfully written to '{file_path}'.")
    except IOError as e:
        print(f"Error writing to file '{file_path}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
my_numbers = [10, 25, 5, 42, 18, 99]
file_to_write = "numbers_list.txt"

write_numbers_to_file(file_to_write, my_numbers)

# Optional: Verify the content of the created file
try:
    with open(file_to_write, 'r') as file:
        print("\nContent of the created file:")
        print(file.read())
except FileNotFoundError:
    print(f"Error: The file '{file_to_write}' was not found after writing.")
```

```python
'''#Q.18.How would you implement a basic logging setup that logs to a file with rotation after 1MB?'''
#Ans.18.
import logging
from logging.handlers import RotatingFileHandler
import os

# Define the log file path
log_file = "rotating_log.log"

# Configure the logger
# Create a logger instance
logger = logging.getLogger("my_rotating_logger")
logger.setLevel(logging.INFO) # Set the minimum logging level

# Create a rotating file handler
# filename: the name of the log file
# maxBytes: the maximum size of the file before rotation (1MB = 1024 * 1024 bytes)
# backupCount: the number of backup files to keep
handler = RotatingFileHandler(log_file, maxBytes=1024 * 1024, backupCount=5)

# Create a formatter
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# Add the formatter to the handler
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# --- Example Usage ---

print(f"Logging to '{log_file}' with rotation at 1MB.")
print("Logging some messages...")

# Log some messages
for i in range(5000): # Log enough messages to potentially trigger rotation
    logger.info(f"This is informational message number {i}")
    if i % 1000 == 0:
        print(f"Logged {i} messages...")

print("\nLogging complete. Check the log file(s).")
print(f"Log file path: {os.path.abspath(log_file)}")
```

```python
'''#Q.19.Write a program that handles both IndexError and KeyError using a try-except block.'''
#Ans.19.
def handle_errors(data, key_to_access, index_to_access):
    """
    Attempts to access a dictionary key and a list index,
    handling KeyError and IndexError exceptions.
    """
    try:
        # Attempt to access a dictionary key
        value_from_dict = data[key_to_access]
        print(f"Value from dictionary for key '{key_to_access}': {value_from_dict}")

        # Attempt to access a list index
        value_from_list = data['my_list'][index_to_access]
        print(f"Value from list at index {index_to_access}: {value_from_list}")

    except (KeyError, IndexError) as e:
        # This block catches both KeyError and IndexError
        print(f"An error occurred: {e}")
        print("This could be due to a missing dictionary key or an invalid list index.")
    except Exception as e:
        # This block catches any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")

# Example usage:
my_data = {
    'name': 'Alice',
    'age': 30,
    'city': 'New York',
    'my_list': [10, 20, 30]
}

print("--- Scenario 1: No errors ---")
handle_errors(my_data, 'name', 1)

print("\n--- Scenario 2: KeyError ---")
handle_errors(my_data, 'country', 0) # 'country' key does not exist

print("\n--- Scenario 3: IndexError ---")
handle_errors(my_data, 'age', 5) # Index 5 is out of bounds for 'my_list'

print("\n--- Scenario 4: Both errors (first encountered is handled) ---")
# In this case, KeyError for 'non_existent_key' will be raised first and handled.
handle_errors(my_data, 'non_existent_key', 10)
```

```python
'''#Q.20.How would you open a file and read its contents using a context manager in Python?'''
#Ans.20.
def read_file_with_context_manager(file_path):
    """
    Opens and reads a file using a context manager.

    Args:
        file_path (str): The path to the file to read.
    """
    try:
        # The 'with' statement ensures the file is closed automatically
        with open(file_path, 'r') as file:
            content = file.read()
            print(f"Content of '{file_path}':\n{content}")
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError as e:
        print(f"An I/O error occurred while accessing '{file_path}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
# Create a dummy file for demonstration
with open("example_context_manager.txt", "w") as f:
    f.write("This is a file created for context manager example.\n")
    f.write("It demonstrates automatic file closing.")

print("--- Reading file using context manager ---")
read_file_with_context_manager("example_context_manager.txt")

print("\n--- Attempting to read a non-existent file ---")
read_file_with_context_manager("non_existent_file_cm.txt")
```

```python
'''#Q.21. Write a Python program that reads a file and prints the number of occurrences of a specific word.'''
#Ans.21.
def count_word_occurrences(file_path, word_to_find):
    """
    Reads a file and counts the occurrences of a specific word (case-insensitive).

    Args:
        file_path (str): The path to the file to read.
        word_to_find (str): The word to count occurrences of.

    Returns:
        int: The number of occurrences of the word, or -1 if the file is not found.
    """
    count = 0
    try:
        with open(file_path, 'r') as file:
            # Read the entire content and convert to lowercase for case-insensitive matching
            content = file.read().lower()
            # Split the content into words and count the occurrences of the target word
            words = content.split()
            count = words.count(word_to_find.lower())
        return count
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        return -1
    except IOError as e:
        print(f"An I/O error occurred while accessing '{file_path}': {e}")
        return -1
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return -1

# Example usage:
# Create a dummy file for demonstration
with open("sample_text.txt", "w") as f:
    f.write("This is a sample text file.\n")
    f.write("This file contains the word sample multiple times.\n")
    f.write("Sample is a good word to sample.")

file_to_read = "sample_text.txt"
word_to_count = "sample"

occurrences = count_word_occurrences(file_to_read, word_to_count)

if occurrences != -1:
    print(f"\nThe word '{word_to_count}' appears {occurrences} times in '{file_to_read}'.")

# Example with a non-existent file
print("\n--- Attempting to count word in a non-existent file ---")
count_word_occurrences("non_existent_file.txt", "test")
```

```python
'''Q.22.How can you check if a file is empty before attempting to read its contents?'''
#Ans.22.


import os

def is_file_empty(file_path):
    """
    Checks if a file exists and is empty.

    Args:
        file_path (str): The path to the file.

    Returns:
        bool: True if the file exists and is empty, False otherwise.
    """
    # Check if the file exists
    if not os.path.exists(file_path):
        print(f"Error: The file '{file_path}' was not found.")
        return False  # File doesn't exist, so it's not empty

    # Check the file size
    if os.path.getsize(file_path) == 0:
        print(f"The file '{file_path}' is empty.")
        return True
    else:
        print(f"The file '{file_path}' is not empty.")
        return False

# Example usage:
# Create a dummy empty file
with open("empty_file_check.txt", "w") as f:
    pass # Creates an empty file

# Create a dummy file with content
with open("non_empty_file_check.txt", "w") as f:
    f.write("This file has content.")

print("--- Checking empty file ---")
is_file_empty("empty_file_check.txt")

print("\n--- Checking non-empty file ---")
is_file_empty("non_empty_file_check.txt")

print("\n--- Checking non-existent file ---")
is_file_empty("non_existent_file_check.txt")

# Clean up the created files (optional)
# os.remove("empty_file_check.txt")
# os.remove("non_empty_file_check.txt")
```

```python
'''#Q.23.Write a Python program that writes to a log file when an error occurs during file handling.'''
#Ans.23.


import logging

# Configure logging to write to a file
logging.basicConfig(filename='file_error.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def read_file_with_error_logging(file_path):
    """
    Attempts to read a file and logs an error if a FileNotFoundError occurs.
    """
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(f"Successfully read file: {file_path}")
            return content
    except FileNotFoundError:
        error_message = f"Error: File not found - {file_path}"
        logging.error(error_message)  # Log the error
        print(error_message)
        return None
    except IOError as e:
        error_message = f"Error reading file {file_path}: {e}"
        logging.error(error_message) # Log other I/O errors
        print(error_message)
        return None
    except Exception as e:
        error_message = f"An unexpected error occurred while handling file {file_path}: {e}"
        logging.error(error_message) # Log any other unexpected errors
        print(error_message)
        return None

# Example usage:
# Attempt to read a file that does not exist
print("--- Attempting to read a non-existent file ---")
read_file_with_error_logging("non_existent_file_for_logging.txt")

# Create a dummy file and attempt to read it (should succeed)
print("\n--- Attempting to read an existing file ---")
with open("existing_file_for_logging.txt", "w") as f:
    f.write("This file exists.")
read_file_with_error_logging("existing_file_for_logging.txt")

print("\nCheck the 'file_error.log' file for logged errors.")
```
